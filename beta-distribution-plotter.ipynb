```python
from bokeh.layouts import column, row
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, Button, TextInput
import numpy as np
import scipy.stats as spst

from bokeh.io import output_notebook, show
output_notebook()
```

```python
def get_data(alpha, beta):
    x = np.linspace(0, .04, 1001)
    y = spst.beta(x, alpha, beta)
    return dict(x=x, y=y)


def modify_doc(doc):
    inp_alpha0 = TextInput(value='1.1', title="alpha_0")
    button_plus_alpha0 = Button(label="teaser 0 clicked", button_type="success")
    inp_alpha1 = TextInput(value='1.2', title="alpha_1")
    button_plus_alpha1 = Button(label="teaser 1 clicked", button_type="success")
    inp_beta0 = TextInput(value='100.0', title="beta_0")
    button_plus_beta0 = Button(label="teaser 0 not clicked", button_type="success")
    button_plus100_beta0 = Button(label="teaser 1 100x not clicked", button_type="success")
    inp_beta1 = TextInput(value='200.0', title="beta_1")
    button_plus_beta1 = Button(label="teaser 1 not clicked", button_type="success")
    button_plus100_beta1 = Button(label="teaser 1 100x not clicked", button_type="success")

    
    alpha0 = float(inp_alpha0.value)
    beta0 = float(inp_beta0.value)
    alpha1 = float(inp_alpha1.value)
    beta1 = float(inp_beta1.value)
    
    source0 = ColumnDataSource(data=get_data(alpha0, beta0))
    source1 = ColumnDataSource(data=get_data(alpha1, beta1))

    p = figure(plot_width=900, plot_height=500, tools="", toolbar_location=None, x_range=(0.0, .04),
               y_range=(0.00001, 500.0))
    p.xaxis.axis_label = 'Teaser CTR'
    p.yaxis.axis_label = 'p(CTR)'
    # log scale
#     p = figure(plot_width=900, plot_height=500, tools="", toolbar_location=None, x_range=(0.0, .04),
#                y_range=(0.00001, 500.0), y_axis_type='log')
    
    r0 = p.line(x='x', y='y', source=source0, line_width=3)
    r1 = p.line(x='x', y='y', source=source1, line_width=3, line_color='orange')

    def update_dist(attrname, old, new):
        alpha0 = float(inp_alpha0.value)
        beta0 = float(inp_beta0.value)
        alpha1 = float(inp_alpha1.value)
        beta1 = float(inp_beta1.value)
        source0.data = get_data(alpha0, beta0)
        source1.data = get_data(alpha1, beta1)
        
    def add_alpha_0(event):
        alpha0 = float(inp_alpha0.value)
        alpha0 += 1.0
        inp_alpha0.value = str(alpha0)
        beta0 = float(inp_beta0.value)
        source0.data = get_data(alpha0, beta0)
        
    def add_alpha_1(event):
        alpha1 = float(inp_alpha1.value)
        alpha1 += 1.0
        inp_alpha1.value = str(alpha1)
        beta1 = float(inp_beta1.value)
        source1.data = get_data(alpha1, beta1)
        
    def add_beta_0(event):
        beta0 = float(inp_beta0.value)
        beta0 += 1.0
        inp_beta0.value = str(beta0)
        alpha0 = float(inp_alpha0.value)
        source0.data = get_data(alpha0, beta0)
        
    def add100_beta_0(event):
        beta0 = float(inp_beta0.value)
        beta0 += 100.0
        inp_beta0.value = str(beta0)
        alpha0 = float(inp_alpha0.value)
        source0.data = get_data(alpha0, beta0)
        
    def add_beta_1(event):
        beta1 = float(inp_beta1.value)
        beta1 += 1.0
        inp_beta1.value = str(beta1)
        alpha1 = float(inp_alpha1.value)
        source1.data = get_data(alpha1, beta1)
        
    def add100_beta_1(event):
        beta1 = float(inp_beta1.value)
        beta1 += 100.0
        inp_beta1.value = str(beta1)
        alpha1 = float(inp_alpha1.value)
        source1.data = get_data(alpha1, beta1)
        
        
    inp_alpha0.on_change('value', update_dist)
    inp_beta0.on_change('value', update_dist)
    inp_alpha1.on_change('value', update_dist)
    inp_beta1.on_change('value', update_dist)
    
    button_plus_alpha0.on_click(add_alpha_0)
    button_plus_alpha1.on_click(add_alpha_1)
    button_plus_beta0.on_click(add_beta_0)
    button_plus100_beta0.on_click(add100_beta_0)
    button_plus_beta1.on_click(add_beta_1)
    button_plus100_beta1.on_click(add100_beta_1)

    layout = column(row(inp_alpha0, inp_beta0, width=800),
                    row(button_plus_alpha0, button_plus_beta0, button_plus100_beta0, width=800),
                    row(inp_alpha1, inp_beta1, width=800),
                    row(button_plus_alpha1, button_plus_beta1, button_plus100_beta1, width=800), row(p))
    
    doc.add_root(layout)

# html = file_html(p, CDN, "my_plot")
show(modify_doc)
```
